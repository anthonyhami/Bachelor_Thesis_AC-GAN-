```python
import pandas as pd
import numpy as np


df_counts = pd.read_csv("Dx_counts_families.csv", sep=";")
print ("df_counts.shape=", df_counts.shape) 


df_counts_rows, df_counts_columns = df_counts.shape
```

```python
df_counts
```

```python
df_counts_sorted = df_counts.sort_values(by=["Dx_counts"])
df_counts_sorted
```

```python
df_counts_sorted = df_counts.sort_values(by=["Dx_counts"])
df_counts_sorted
```

```python

df_summary_evaluation = pd.read_csv("df_summary_evaluation.csv", sep=";")
print ("df_summary_evaluation.shape=", df_summary_evaluation.shape) 


df_summary_evaluation_rows, df_summary_evaluation_columns = df_summary_evaluation.shape
```

```python
df_summary_evaluation
```

```python
df_summary_evaluation
```

```python
def median(df, col_name="number_of_nearest_classes"):
    return df[col_name].median()

median_of_all_classes = median(df_summary_evaluation, col_name="number_of_nearest_classes")
median_of_all_classes  
```

```python
df_summary_evaluation_centroids = df_summary_evaluation[df_summary_evaluation['fake_index'] == 11] 
df_summary_evaluation_centroids
```

```python
median_of_centroids = median(df_summary_evaluation_centroids, col_name="number_of_nearest_classes")
median_of_centroids
```

```python
df_grouped = df_summary_evaluation.groupby("class_index")

for name, group in df_grouped:
    real_name = np.unique(group["real_name"])[0]
    print ("real_name ==== ", real_name)
    gr = group.iloc[:10, :] # do not take the centroid
    median_of_each_class = median(gr, col_name="number_of_nearest_classes")
    print ("median_of_each_class == ", median_of_each_class)
```

```python
df_summary_evaluation_no_centroids = df_summary_evaluation[df_summary_evaluation['fake_index'] != 11] 

median_no_centroids = median(df_summary_evaluation_no_centroids, col_name="number_of_nearest_classes")
print ("median_no_centroids === ", median_no_centroids)
df_summary_evaluation_no_centroids
```

```python
low_than_20_values = df_counts_sorted[df_counts_sorted["Dx_counts"] < 20]["Dx"].values

df_summary_evaluation_low_than_20_values = df_summary_evaluation[df_summary_evaluation['real_name'].isin(low_than_20_values)] 
median_low_than_20_values = median(df_summary_evaluation_low_than_20_values, col_name="number_of_nearest_classes")
print ("median_low_than_20_values == ", median_low_than_20_values)
df_summary_evaluation_low_than_20_values
```

```python
higher_than_100_values = df_counts_sorted[df_counts_sorted["Dx_counts"] > 100]["Dx"].values

df_summary_evaluation_higher_than_100_values = df_summary_evaluation[df_summary_evaluation['real_name'].isin(higher_than_100_values)] 
median_higher_than_100_values = median(df_summary_evaluation_higher_than_100_values, col_name="number_of_nearest_classes")
print ("median_higher_than_100_values == ", median_higher_than_100_values)
df_summary_evaluation_higher_than_100_values
```
